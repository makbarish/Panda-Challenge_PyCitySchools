```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
Total_Schools = school_data_complete['School ID'].nunique()
Total_Students = school_data_complete['Student ID'].nunique()
Total_Budget = school_data['budget'].sum()

Avereage_Math_Scores = school_data_complete['math_score'].mean()
Average_Reading_Score = school_data_complete['reading_score'].mean()

Passing_Math = school_data_complete.loc[school_data_complete['math_score'] >= 70, ['math_score']].count()
Passing_Math_Percentage = Passing_Math / Total_Students * 100
    
Passing_Reading = school_data_complete.loc[school_data_complete['reading_score'] >= 70,['reading_score']].count()
Passing_Reading_Percentage = Passing_Reading / Total_Students * 100

Overal_Passing = student_data[(student_data['math_score'] >= 70) & (student_data['reading_score'] >= 70)].count()
Overal_Passing_Percentage = Overal_Passing / Total_Students * 100

dist_summary_df = pd.DataFrame({  
                                "Total Schools": Total_Schools,
                                "Total Students": Total_Students,
                                "Total Budget": Total_Budget,
                                "Avereage Math Scores": Avereage_Math_Scores,
                                "Average Reading Score": Average_Reading_Score,
                                "% Passing Math": Passing_Math_Percentage,
                                "% Passing Reading": Passing_Reading_Percentage,
                                "% Overal Passing": Overal_Passing_Percentage
                               })

dist_summary_df["Total Schools"] = dist_summary_df["Total Schools"].map("{:,}".format)
dist_summary_df["Total Students"] = dist_summary_df["Total Students"].map("{:,}".format)
dist_summary_df["Total Budget"] = dist_summary_df["Total Budget"].map("${:,}".format)
dist_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
total_schools = school_data_complete['school_name'].value_counts()

grouped_schools = school_data_complete.groupby('school_name')
school_budget = grouped_schools['budget'].unique()

per_student_budget = school_budget/total_schools

school_type = grouped_schools["type"].unique()
school_ave_Math = grouped_schools['math_score'].mean()
school_ave_Reading = grouped_schools['reading_score'].mean()

# Passing grades are 70 and above
school_Pass_Math = school_data_complete.loc[school_data_complete['math_score'] >= 70] 
school_Pass_Math_Grp = school_Pass_Math.groupby(['school_name']).count()
math_Percent_Pass = (school_Pass_Math_Grp['math_score']/total_schools)*100

school_Pass_Reading_Grp = school_Pass_Reading.groupby(['school_name']).count()
school_Pass_Reading = school_data_complete.loc[school_data_complete['reading_score'] >= 70] 
reading_Percent_Pass = (school_Pass_Reading_Grp['reading_score']/total_schools)*100

overal_passing = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)].groupby('school_name')['Student ID'].count()
Overal_Passing_Percentage = Overal_Passing / total_schools * 100
    
school_summary = pd.DataFrame({
    
                            "School Type": school_type,
                            "Total Students":total_schools,
                            "Total School Budget": school_budget, 
                            "Per Student Budget": per_student_budget,
                            "Average Math Score": school_ave_Math, 
                            "Average Reading Score":school_ave_Reading, 
                            "% Passing Math":math_Percent_Pass, 
                            "% Passing Reading":reading_Percent_Pass,
                            "% Overall Passing":Overal_Passing_Percentage

                            })


school_summary["Total Students"] = school_summary["Total Students"].map("{:,}".format)
# school_summary["Per Student Budget"] = school_summary["Per Student Budget"].map("${:,}".format)
# school_summary["Total School Budget"] = school_summary["Total School Budget"].map("{:,}".format)

school_summary
```
